```python
import pandas as pd
import numpy as np
import scipy.stats as stat
import scipy as sp
import statistics
df = pd.read_csv("df_33.csv",index_col=0)
df.head(2)
```

```python
df= df.drop(["Ville", "Region", "Departement"], axis = 1)
```

```python
X =df.drop(["Succes"], axis = 1)
```

```python
y = df['Succes']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import lightgbm as lgb
```

```python
clf = lgb.LGBMClassifier()
scaler = StandardScaler()
pipe_lgbm = make_pipeline(scaler, lgb.LGBMClassifier())
pipe_lgbm.fit(X_train, y_train)
```

```python
y_clf = pipe_lgbm.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_clf,target_names = ['échec(Class 0)','succès (Class 1)']))
```
