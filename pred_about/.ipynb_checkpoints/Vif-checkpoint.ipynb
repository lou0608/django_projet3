```python
df = pd.read_csv("df.csv")
```

```python
df=df.drop(columns="Unnamed: 0", axis = 1)
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Applying scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['Code', 'Discriminator', 'PrixAchatReel', 'TypeDeMission', 'TypeBienEnum','TypeProjet', 'BudgetMinEuro','BudgetMaxEuro', 'SurfaceMin', 'NombrePiecesEnum', 'NombreChambresEnum', 'DescriptionBien_Prix', 'Statut']
df[num_vars] = scaler.fit_transform(df[num_vars])

df
```

```python
# x, mes variables independantes
x = df[["Code", "Discriminator", "PrixAchatReel","TypeDeMission","TypeBienEnum","TypeProjet","BudgetMinEuro", "BudgetMaxEuro", "SurfaceMin", "NombrePiecesEnum", "NombreChambresEnum", "AncienNeufBien", "DescriptionBien_TypeBienEnum", "DescriptionBien_Prix", "Statut"]]
x.shape
```

```python
# y, ma varible dependante
y = df["DureeMandat"]
y.shape
```

```python
# separation du jeu d'entrainement et du jeu de test
# jeu d'entrainement / apprentissage du modèle
# jeu de test == test du modèle
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 100 )
```

```python
#Build a linear model

import statsmodels.api as sm
x_train_lm = sm.add_constant(x_train)

lr_1 = sm.OLS(y_train, x_train_lm).fit()

lr_1.summary()
```

```python
# Checking for the VIF values of the variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Creating a dataframe that will contain the names of all the feature variables and their VIFs
vif = pd.DataFrame()
vif['Features'] = x_train.columns
vif['VIF'] = [variance_inflation_factor(x_train.values, i) for i in range(x_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
x = x_train.drop('BudgetMinEuro', 1,)

# Build a fitted model after dropping the variable
x_train_lm = sm.add_constant(x)

lr_2 = sm.OLS(y_train, x_train_lm).fit()

# Printing the summary of the model
print(lr_2.summary())
```

```python
# Calculating the VIFs again for the new model after dropping semi-furnished

vif = pd.DataFrame()
vif['Features'] = x.columns
vif['VIF'] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
